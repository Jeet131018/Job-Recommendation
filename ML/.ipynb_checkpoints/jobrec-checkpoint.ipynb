```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MyDataSet2.csv")
df = df.drop(df.columns[:1], axis=1)
df
```

```python
df.dropna(subset=['url'], inplace=True)
df.head(20)
```

```python
df['skills']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
df['skills']=df['skills'].fillna('')
tfidf_matrix =tfidf.fit_transform(df['skills'])
#tfidf_matrix.shape
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df.index, index=df['search_type']).drop_duplicates()
```

```python
def get_recommendations(search_type, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[search_type]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    job_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return (df.iloc[job_indices])
```

```python
get_recommendations('JAVA')
```
